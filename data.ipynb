```python
# IMports
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Cell 2: Load the CSV file into a DataFrame
df = pd.read_csv('data.csv')
```

```python
# Display the first few rows to confirm loading
df.head()
```

```python
# Cell 3: Basic information about the dataset
print("--- Dataset Info ---")
df.info() # Data types, non-null counts, memory usage
print("\n--- Dataset Shape (Rows, Columns) ---")
print(df.shape)
print("\n--- Column Names ---")
print(df.columns.tolist())
```

```python
# Drop the specified column in place(modifies the original DataFrame)
df = df.drop(columns=['Net foreign assets (current LCU) ttd'])
df = df.drop(columns=['Year'])

# Verify the column has been dropped by checking the columns again
print(df.columns.tolist())
```

```python
print("--- Statistical Summary (Numerical Columns) ---")
df.describe()
```

```python
print("--- Correlation Matrix (Numerical Columns) ---")
correlation_matrix = df.select_dtypes(include='number').corr()
print(correlation_matrix)
```

```python
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.show()
```
